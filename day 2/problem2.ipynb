```python
import pandas as pd
```

```python
listas = pd.read_csv('relagorios.csv')
```

```python
def verifica_ordem(linha):
    valor = None
    diferenca = None
    valido = True

    print("_________________________________")

    print("0_____linha", linha)

    #for item in linha.split():
    for item in linha.split():
        print("1____item",item)

        if not valido:
            break

        item = int(item)
        if valor:
            diferenca_nova = item - valor

            print("2____diferenca_nova",diferenca_nova)

            valido = 4 > abs(diferenca_nova) > 0

            print("3____novo_valido",valido)

            if valido and diferenca_nova is not None and diferenca is not None :
                valido = ((diferenca < 0 and diferenca_nova < 0) or (diferenca > 0 and diferenca_nova > 0))
                print("4____entrou no if", valido)

                print("6____ja tinha removido antes ou esta valido", diferenca_nova)
            diferenca = diferenca_nova
        valor = item


    print("7____end status", valido)
    return valido


```

```python
listas["valido"] = listas.apply(verifica_ordem, axis=1)
```

```python
listas["valido"].value_counts()
```

```python
listas_invalidas = listas[listas["valido"]==False].copy()
```

```python
def gerar_listas(row):
    valores = row.split()
    listas_resultantes = [" ".join(valores[:i] + valores[i+1:]) for i in range(len(valores))]
    return listas_resultantes
```

```python
listas_invalidas["listas"] = listas_invalidas["relatorios"].apply(gerar_listas)
```

```python
listas_invalidas
```

```python
for index, item in listas_invalidas.iterrows():
    print(item)
    valido = False
    for opcao in item["listas"]:
        if verifica_ordem(opcao):
            valido = True
            listas_invalidas.loc[index, 'valido'] = valido
            break
```
