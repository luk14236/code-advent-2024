```python
def parse_robot(line):
    # Exemplo de linha: "p=0,4 v=3,-3"
    # Vamos extrair p e v.
    line = line.strip()
    parts = line.split()
    p_part = parts[0]  # p=0,4
    v_part = parts[1]  # v=3,-3

    # Remover 'p=' e 'v=' e depois dividir por ','
    px, py = p_part[2:].split(',')
    vx, vy = v_part[2:].split(',')

    px, py = int(px), int(py)
    vx, vy = int(vx), int(vy)

    return (px, py, vx, vy)

def print_grid(positions, width, height, arquivo_nome):
    # Agora o print_grid espera somente (x, y)
    grid = [['.' for _ in range(width)] for _ in range(height)]
    resultado = ""
    for (x, y) in positions:
        grid[y][x] = '#'
    for row in grid:
        resultado = resultado + "".join(row)

    filename = f"{arquivo_nome}.txt"

    with open(filename, 'w') as f:
     f.write(resultado)

    print("\n")  # Uma linha em branco após

def simulate(robots, width, height, min_time, max_time):
    while min_time <= max_time:
        # Simula o movimento dos robôs por 'time' segundos.
        # Simula o movimento dos robôs por 'time' segundos.
        final_positions = []
        for (px, py, vx, vy) in robots:
            final_x = (px + vx * min_time) % width
            final_y = (py + vy * min_time) % height
            final_positions.append((final_x, final_y))
        # Chama print_grid com a lista final de posições
        print_grid(final_positions, width, height, min_time)

        min_time += 1
    return final_positions

def count_quadrants(positions, width=101, height=103):
    # Linhas centrais
    center_x = width // 2      # 101//2 = 50
    center_y = height // 2     # 103//2 = 51

    top_left = 0
    top_right = 0
    bottom_left = 0
    bottom_right = 0

    for (x, y) in positions:
        # Se estiver exatamente na linha central, não conta
        if x == center_x or y == center_y:
            continue
        elif x < center_x and y < center_y:
            top_left += 1
        elif x > center_x and y < center_y:
            top_right += 1
        elif x < center_x and y > center_y:
            bottom_left += 1
        elif x > center_x and y > center_y:
            bottom_right += 1

    return top_left, top_right, bottom_left, bottom_right

# Exemplo de uso:
if __name__ == "__main__":
    # Supondo que input_lines_text esteja definido em algum lugar
    input_lines = input_lines_text.split("\n")

    robots = [parse_robot(line) for line in input_lines if line.strip()]
    final_positions = simulate(robots, width=101, height=103, min_time=7050, max_time=7060)
    # Ajustar a altura de volta para 103, conforme o enunciado original
    tl, tr, bl, br = count_quadrants(final_positions, width=101, height=103)
    safety_factor = tl * tr * bl * br
    print("Fator de segurança:", safety_factor)
```
