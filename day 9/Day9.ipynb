```python
def compact_disk(texto):
    # Transformar o input denso em uma representação detalhada
    disk = []
    file_id = 0

    for i in range(0, len(texto), 2):
        file_length = int(texto[i])
        free_space_length = int(texto[i + 1]) if i + 1 < len(texto) else 0

        if file_length > 0:
            disk.extend([file_id] * file_length)
            file_id += 1

        if free_space_length > 0:
            disk.extend(["."] * free_space_length)

    print (disk)

    # Compactar os arquivos no disco
    while "." in disk:
        # Encontra o índice mais à direita de um bloco de arquivo
        for i in range(len(disk) - 1, -1, -1):
            if disk[i] != ".":
                break

        # Encontra o índice mais à esquerda do espaço livre
        for j in range(len(disk)):
            if disk[j] == ".":
                break

        # Se não houver mais espaço para mover, pare
        if i < j:
            break

        # Move o bloco de arquivo
        disk[j], disk[i] = disk[i], "."

    # Calcular o checksum
    checksum = sum(pos * file_id for pos, file_id in enumerate(disk) if file_id != ".")

    return checksum

# Exemplo de uso com o input do desafio
texto = "3364353585152525276734103158952168605835658851512146272137997021256238991376519842415284776"
checksum_result = compact_disk(texto)
print("Checksum resultante:", checksum_result)
```

```python
from collections import deque

# Função para calcular o checksum
def calculate_checksum(disk_map):
    checksum = 0
    for i, char in enumerate(disk_map):
        if char.isdigit():
            checksum += i * int(char)
    return checksum

# Função para compactar o disco
def compact_disk(disk_map):
    # Transformar o mapa em uma lista para manipulação
    disk_map = list(disk_map)
    n = len(disk_map)

    # Manter um deque para rastrear os índices dos espaços vazios
    free_space_indices = deque(i for i, char in enumerate(disk_map) if char == '.')

    # Mover blocos de arquivos para a esquerda
    for i in range(n):
        if free_space_indices and disk_map[i].isdigit():
            free_index = free_space_indices.popleft()  # Pega o primeiro espaço vazio
            if free_index < i:  # O espaço vazio está à esquerda do bloco
                disk_map[free_index], disk_map[i] = disk_map[i], '.'
                free_space_indices.append(i)  # Atualiza a posição do espaço vazio

    return ''.join(disk_map)

# Entrada do problema
input_disk_map = texto

# Criar o mapa expandido
expanded_map = []
file_id = 0

# Processar os pares
for i in range(0, len(input_disk_map), 2):
    file_blocks = int(input_disk_map[i])  # Número de blocos do arquivo
    free_blocks = int(input_disk_map[i+1])  # Número de blocos livres
    expanded_map.extend([str(file_id)] * file_blocks)
    expanded_map.extend(['.'] * free_blocks)
    file_id += 1

# Convertendo o mapa expandido para string
expanded_map = ''.join(expanded_map)

# Compactar o disco (a função compact_disk deve ser usada aqui)
compacted_map = compact_disk(expanded_map)

# Calcular o checksum
checksum = calculate_checksum(compacted_map)

print("Mapa do disco original:", expanded_map)
print("Mapa do disco compactado:", compacted_map)
print("Checksum do sistema de arquivos:", checksum)
```
