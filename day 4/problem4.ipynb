```python
def find_all_a_positions(matrix):
    positions = []
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            if matrix[i][j] == 'A':
                positions.append((i, j))
    return positions

def is_xmas_in_x(matrix, row, col):
    try:
        diag1_valid = (
            (matrix[row-1][col-1] == 'M' and matrix[row+1][col+1] == 'S') or
            (matrix[row-1][col-1] == 'S' and matrix[row+1][col+1] == 'M')
        )
        diag2_valid = (
            (matrix[row-1][col+1] == 'M' and matrix[row+1][col-1] == 'S') or
            (matrix[row-1][col+1] == 'S' and matrix[row+1][col-1] == 'M')
        )
        return diag1_valid and diag2_valid
    except IndexError:
        return False

def find_xmas_positions(matrix):
    a_positions = find_all_a_positions(matrix)
    xmas_positions = []
    for row, col in a_positions:
        if is_xmas_in_x(matrix, row, col):
            xmas_positions.append((row, col))
    return xmas_positions

xmas_positions = find_xmas_positions(matriz_texto)
len(xmas_positions)
```
