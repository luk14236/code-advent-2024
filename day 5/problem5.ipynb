```python
def lista_relevantes(manual):

    regras_relevantes = []

    for regra in regras:
        regra_splited = regra.split('|')

        try:
            if regra_splited[0] in manual and regra_splited[1] in manual:
                regras_relevantes.append(regra_splited)
        except:
            continue

    return regras_relevantes
```

```python
def verifica_regra(manual):
    for regra in lista_relevantes(manual):
        valido = manual.index(regra[0]) < manual.index(regra[1])

        if not valido:
            return False

    return manual
```

```python
from collections import defaultdict, deque

def ordenar_e_validar_updates(regras, invalid_list):
    novos_validos = 0

    # Criar o grafo de dependncias
    grafo = defaultdict(list)
    in_degree = defaultdict(int)

    for regra in regras:
        x, y = regra.split('|')
        grafo[x].append(y)
        in_degree[y] += 1
        if x not in in_degree:
            in_degree[x] = 0

    for invalido in invalid_list:
        print("1___Iniciando a validao do ->", invalido)
        elementos = invalido.split(',')
        presente = set(elementos)

        subgrafo = defaultdict(list)
        sub_in_degree = defaultdict(int)
        for pagina in elementos:
            sub_in_degree[pagina] = 0

        for x in presente:
            for y in grafo[x]:
                if y in presente:
                    subgrafo[x].append(y)
                    sub_in_degree[y] += 1

        fila = deque([node for node in elementos if sub_in_degree[node] == 0])
        ordenado = []

        while fila:
            atual = fila.popleft()
            ordenado.append(atual)
            for vizinho in subgrafo[atual]:
                sub_in_degree[vizinho] -= 1
                if sub_in_degree[vizinho] == 0:
                    fila.append(vizinho)


        if len(ordenado) != len(elementos):
            print("Erro: ciclo detectado ou ordenao incompleta.", ordenado)
            continue

        # Soma do elemento central
        manual_list = ordenado
        print("Permutao vlida encontrada ->", ','.join(manual_list))
        novos_validos += int(manual_list[len(manual_list)//2])  # Soma o elemento central

    print("3___Novos vlidos ->", novos_validos)
    return novos_validos


novos_validos = ordenar_e_validar_updates(regras, invalid_list)
print("Resultado final:", novos_validos)
```
