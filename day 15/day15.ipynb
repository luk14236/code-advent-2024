```python
def expandir_mapa(mapa):
    mapa_expandido = []
    for fila in mapa:
        nova_fila = ""
        for c in fila:
            if c == 'O':
                nova_fila += '[]'
            elif c == '@':
                nova_fila += '@.'
            elif c == '#':
                nova_fila += '##'
            elif c == '.':
                nova_fila += '..'
            else:
                # Caso apareça algo inesperado
                nova_fila += c
        mapa_expandido.append(nova_fila)
    return mapa_expandido

def read_input():
    # As variáveis input_lines e input_moves já devem estar definidas externamente
    map_lines = input_lines.splitlines()
    map_lines = expandir_mapa(map_lines)

    move_lines = "".join(input_moves.split())
    moves = list(move_lines)

    return map_lines, moves

def find_robot_and_boxes(warehouse):
    robot_pos = None
    boxes = set()
    for y, row in enumerate(warehouse):
        for x, cell in enumerate(row):
            if cell == '@.':
                robot_pos = (x, y)
            elif cell == '[]':
                boxes.add((x, y))
    return robot_pos, boxes

def direction_delta(d):
    if d == '^':
        return (0, -1)
    elif d == 'v':
        return (0, 1)
    elif d == '<':
        return (-1, 0)
    elif d == '>':
        return (1, 0)
    return (0, 0)

def can_move(robot, boxes, dx, dy, warehouse):
    x, y = robot
    nx, ny = x + dx, y + dy

    # Verifica limites
    if ny < 0 or ny >= len(warehouse) or nx < 0 or nx >= len(warehouse[ny]):
        return False, robot, boxes

    # Posição destino é parede?
    if warehouse[ny][nx] == '##':
        return False, robot, boxes

    # Se não é caixa, move o robô
    if (nx, ny) not in boxes:
        return True, (nx, ny), boxes

    # Empurrar cadeia de caixas
    chain = []
    cx, cy = nx, ny
    while (cx, cy) in boxes:
        chain.append((cx, cy))
        cx += dx
        cy += dy

    # Verifica espaço após a cadeia
    if cy < 0 or cy >= len(warehouse) or cx < 0 or cx >= len(warehouse[cy]):
        return False, robot, boxes
    if warehouse[cy][cx] == '##':
        return False, robot, boxes

    # Move as caixas
    new_boxes = set(boxes)
    new_boxes -= set(chain)
    for (bx, by) in chain:
        new_boxes.add((bx + dx, by + dy))

    # Move o robô
    return True, (nx, ny), new_boxes

def compute_gps(boxes):
    # GPS = 100*y + 2*x (2*x pois cada tile tem 2 chars de largura)
    return sum(100 * y + 2 * x for (x, y) in boxes)

# Execução
map_lines, moves = read_input()

warehouse = []
for line in map_lines:
    # Divide a linha em blocos de 2 caracteres
    row = [line[i:i+2] for i in range(0, len(line), 2)]
    warehouse.append(row)

robot, boxes = find_robot_and_boxes(warehouse)

for m in moves:
    dx, dy = direction_delta(m)
    moved, new_robot, new_boxes = can_move(robot, boxes, dx, dy, warehouse)
    if moved:
        robot = new_robot
        boxes = new_boxes

result = compute_gps(boxes)
print(result)
```
