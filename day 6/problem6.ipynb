```python
import numpy as np

# Funes auxiliares para movimento do guarda
def move(guarda_posicao, direcao):
    if direcao == "cima":
        return (guarda_posicao[0] - 1, guarda_posicao[1]), "cima"
    elif direcao == "direita":
        return (guarda_posicao[0], guarda_posicao[1] + 1), "direita"
    elif direcao == "desce":
        return (guarda_posicao[0] + 1, guarda_posicao[1]), "desce"
    elif direcao == "esquerda":
        return (guarda_posicao[0], guarda_posicao[1] - 1), "esquerda"
    return guarda_posicao, direcao

def turn_right(direcao):
    return {
        "cima": "direita",
        "direita": "desce",
        "desce": "esquerda",
        "esquerda": "cima"
    }[direcao]

# Simulao do movimento do guarda
def simulate_guard(sala_matriz, guarda_posicao, direcao):
    limite_y, limite_x = sala_matriz.shape
    visited = set()  # Armazena (posio, direo)

    while 0 <= guarda_posicao[0] < limite_y and 0 <= guarda_posicao[1] < limite_x:
        # Verifica se estamos em um ciclo
        state = (guarda_posicao, direcao)
        if state in visited:
            return True  # Loop detectado
        visited.add(state)

        # Move o guarda
        next_pos, next_dir = move(guarda_posicao, direcao)

        # Checa se h obstruo
        if (
            0 <= next_pos[0] < limite_y and 0 <= next_pos[1] < limite_x and
            sala_matriz[next_pos[0], next_pos[1]] == "#"
        ):
            direcao = turn_right(direcao)  # Vira  direita
        else:
            guarda_posicao = next_pos  # Continua em frente

    return False  # Guarda saiu do mapa

# Encontrar posies para obstrues que causam loops
def find_obstruction_positions(sala_matriz, guarda_posicao, direcao):
    limite_y, limite_x = sala_matriz.shape
    possible_positions = []

    for y in range(limite_y):
        for x in range(limite_x):
            # Ignorar a posio inicial e obstrues existentes
            if (y, x) == guarda_posicao or sala_matriz[y, x] == "#":
                continue

            # Testar colocando uma obstruo
            original = sala_matriz[y, x]
            sala_matriz[y, x] = "#"
            if simulate_guard(sala_matriz, guarda_posicao, direcao):
                possible_positions.append((y, x))
            sala_matriz[y, x] = original  # Reverter para o estado original

    return possible_positions

# Exemplo de entrada
sala_matriz = [list(linha) for linha in sala.split('\n')]
sala_matriz = np.array(sala_matriz)

# Posio inicial e direo do guarda
guarda_posicao = (39, 46)  # Linha 6, coluna 4
direcao = "cima"

# Encontrar posies possveis
possiveis_obstrucoes = find_obstruction_positions(sala_matriz, guarda_posicao, direcao)

# Resultado
print("Posies para obstrues que causam loop:", possiveis_obstrucoes)
print("Nmero de posies possveis:", len(possiveis_obstrucoes))
```
